```julia
using PyPlot
using Statistics


x = [1.9223, 2.1234, 1.9324, 3.00123, 2.3143, 2.4365, 1.65438, 
    1.8223, 2.1789, 2.6544, 2.98903, 2.5142, 2.3465, 1.96548]

xbar = mean(x)
σ = std(x)
println(x)
println("mean = ", xbar)
println("σ = ", σ)

inside_σ = filter(i -> i > xbar - σ && i < xbar + σ, x)

println(inside_σ)
println(length(inside_σ) / length(x))

inside_3σ = filter(i -> i > xbar - 3*σ && i < xbar + 3*σ, x)

println(inside_3σ)
println(length(inside_3σ) / length(x))
```

```julia
#Monte Carlo simulation

using PyPlot
using Random
import Statistics

N = 1000000
K0 = 273 # K at 0 °C
# Nominal values
α_nom = 0.0209 # Ω
β_nom = 3900 # K

α = α_nom .+ ((α_nom * 0.02) .* ((rand(N) .* 2) .- 1)) # Ω
β = β_nom .+ ((β_nom * 0.03) .* ((rand(N) .* 2) .- 1)) # °K

R(T) = α .* exp.(β / T) # R in Ω, T in Kelvin
R_nom(T) = α_nom * exp(β_nom / T) # Nominal value of R in Ω at T in K 

for TC in [25, 30] # Temperature in °C
    R_set = R(K0 + TC)
    σ = Statistics.std(R_set)
    R_bar = Statistics.mean(R_set)
    R_med = Statistics.median(R_set)
    figure(); title("MC at $TC °C"); hist(R_set, 30); 
    
    show();
    println("For ", TC, "°C:")
    println("σ = ", σ)
    println("mean(R) = ", R_bar)
    println("median(R) = ", R_med)
    println("Nominal R = ", R_nom(K0 + TC))
end
    
```
